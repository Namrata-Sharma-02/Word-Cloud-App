```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
import requests
from bs4 import BeautifulSoup
#URL = "https://en.wikipedia.org/wiki/Kingdom_of_Jerusalem"
URL = "https://www.analyticsvidhya.com/blog/2021/05/how-to-build-word-cloud-in-python/"
#URL = "https://en.wikipedia.org/wiki/India"
page = requests.get(URL)

soup = BeautifulSoup(page.content,"html.parser")

#print(page.text)
soup.title.text
```

```python
job_elements = soup.find_all("p")
print(job_elements)
```

```python
job_elements[10]
my_string = ""

for item in job_elements:
    my_string = my_string + " "+item.text
    
my_string
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(my_string)
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.rcParams["figure.figsize"] = (15,15)
plt.show()
```
